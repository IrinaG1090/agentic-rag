```python
import os
from dotenv import load_dotenv
from langchain.tools import tool
from langchain_ollama import ChatOllama
from langgraph.prebuilt import create_react_agent

load_dotenv()
print("✅ Импорты готовы")
```

```python
@tool
def calculate(expression: str) -> str:
    """Вычисляет математическое выражение (+, -, *, /, **, sqrt, и т.д.)."""
    import math
    import re
    
    # Безопасно вычисляем выражение
    allowed_names = {k: v for k, v in math.__dict__.items() if not k.startswith("__")}
    allowed_names["abs"] = abs
    
    try:
        # Санитизация ввода
        if re.search(r'[^0-9+\-*/()., eE\s]', expression):
            return "Ошибка: недопустимые символы"
        
        result = eval(expression, {"__builtins__": {}}, allowed_names)
        return f"Результат: {result}"
    except Exception as e:
        return f"Ошибка вычисления: {e}"

print("✅ Калькулятор готов")
```

```python
@tool
def search_documents(query: str) -> str:
    """Ищет информацию в загруженных документах."""
    return f"[Поиск по запросу: {query}]"
print("✅ Поиск готов")
```

```python
tools = [calculate, search_documents]
print("✅ Инструменты:", [t.name for t in tools])
```

```python
llm = ChatOllama(
    model="phi4-mini:3.8b-q4_K_M",
    temperature=0,
    num_thread=2
)
print("✅ Модель готова")
```

```python
agent = create_react_agent(
    model=llm,
    tools=tools
)
print("✅ Агент LangGraph готов")
```

```python
questions = [
    "Сколько будет 25 * 4 + 10?",
    "Что такое in-context learning?",
    "Вычисли 2 в 10 степени"
]

for i, q in enumerate(questions, 1):
    print(f"\n{i}. {q}")
    result = agent.invoke({"messages": [("user", q)]})
    print(f"Ответ: {result['messages'][-1].content}")
```
